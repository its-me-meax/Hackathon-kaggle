```python
## filling Null with Median :->
df.NUMFLOORS.fillna(df.NUMFLOORS.median(), inplace=True)
df.NUMBEDS.fillna(df.NUMBEDS.median(), inplace=True)
df.ExtendedQuantity.fillna(df.ExtendedQuantity.median(), inplace=True)
df.invoiceTotal.fillna(df.invoiceTotal.median(), inplace=True)

#3 filling Null with mode :-> 
# Check if the column is not entirely null before filling
if not df.PriceUOM.isnull().all():
    df.PriceUOM.fillna(df.PriceUOM.mode()[0], inplace=True)

# Repeat for the other columns
if not df.UOM.isnull().all():
    df.UOM.fillna(df.UOM.mode()[0], inplace=True)

if not df.PROJECT_TYPE.isnull().all():
    df.PROJECT_TYPE.fillna(df.PROJECT_TYPE.mode()[0], inplace=True)




## Extended Price :-> 
df['ExtendedPrice'].fillna(0, inplace=True)  # or recompute if formula available

```
